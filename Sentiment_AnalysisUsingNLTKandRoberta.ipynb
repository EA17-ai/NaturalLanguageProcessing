1.Using Vaders Model to perform Sentiment Analysis

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
sla=SentimentIntensityAnalyzer()
```

```python
sla.polarity_scores(text)
```

```python
results={}
for i,row in tqdm(df1.iterrows(),total=len(df1)):
  text_row=row["Text"]
  rowid=row["Id"]
  results[rowid]=sla.polarity_scores(text_row)
```

```python
vaders=pd.DataFrame(results).T
vaders.head(4)
```

```python
vaders=vaders.reset_index().rename(columns={"index":"Id"})
vaders.head(4)
```

```python
vaders_merged=vaders.merge(df1,how="left")
vaders_merged.head(4)
```

```python
sns.barplot(data=vaders_merged,x="Score",y="compound")
```

```python
fig,axs=plt.subplots(1,3,figsize=(15,5))
sns.barplot(data=vaders_merged,x="Score",y="pos",ax=axs[0])
sns.barplot(data=vaders_merged,x="Score",y="neu",ax=axs[1])
sns.barplot(data=vaders_merged,x="Score",y="neg",ax=axs[2])
```

```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
```

```python
Model=f"cardiffnlp/twitter-roberta-base-sentiment"
tokenizer=AutoTokenizer.from_pretrained(Model)
model=AutoModelForSequenceClassification.from_pretained(Model)
```

```python
encoded_text=tokenizer(text,return_tensors="pt")
output=model(**encoded_text)
scores=output[0][0].detach().numpy()
scores=softmax(scores)
scores_dict={"rob_neg":scores[0],
             "rob_neu":scores[1],
             "rob_pos":scores[2]}
print(scores_dict)
```

```python
def polority_score_roberta(example):
  encoded_text=tokenizer(example,return_tensors="pt")
  output=model(**encoded_text)
  scores=output[0][0].detach().numpy()
  scores=softmax(scores)
  scores_dict={"rob_neg":scores[0],
             "rob_neu":scores[1],
             "rob_pos":scores[2]}
  return scores_dict
```

```python
res1={}
for i,row in tqdm(df1.iterrows(),total=len(df1)):
  try:
    myid=row["Id"]
    res1[myid]=polority_score_roberta(row["Text"])
  except RuntimeError:
     print(f"Broke for Id {myid}")
```

```python
res_df=pd.DataFrame(res1).T
```

```python
res_df
```

```python
res_df=res_df.reset_index().rename(columns={"index":"Id"})
res_df.head(4)
```

```python
res_df_merged=res_df.merge(vaders_merged,how="left")
```
